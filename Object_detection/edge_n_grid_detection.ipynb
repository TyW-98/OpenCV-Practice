```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def display_img(img,cmap = None):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread("../DATA/sammy.jpg")
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

display_img(img_rgb)
```

```python
blurred_img = cv2.blur(img,ksize=(7,7))

# Formula to select the starting threshold value
med_val = np.median(img)

# Select whichever is greater either 0 or 70% of the median value
lower_threshold = int(max(0,0.7*med_val))

# Select whichever is the samller either 255 or 130% of the median value
upper_threshold = int((min(255,1.3*med_val)))

img_edge = cv2.Canny(blurred_img,threshold1=lower_threshold,threshold2=upper_threshold)

display_img(img_edge)
```

```python
flat_chess_img = cv2.imread("../DATA/flat_chessboard.png")

display_img(flat_chess_img)
```

```python
# (7,7) = The number of squares - 1 since the edge boxes cant be detected
found , corners = cv2.findChessboardCorners(flat_chess_img,(7,7))

print(found)

print("\n")

print(corners)
```

```python
cv2.drawChessboardCorners(flat_chess_img,(7,7),corners,found)

display_img(flat_chess_img)
```
