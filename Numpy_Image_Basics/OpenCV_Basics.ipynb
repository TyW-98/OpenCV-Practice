```python
image_1_gray = cv2.imread("../../Course-Resource/DATA/00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(image_1_gray,cmap="gray")
image_1_gray.shape
```

```python
image_1_resize = cv2.resize(image_1_rgb,(550,256))
```

```python
plt.imshow(image_1_resize)
```

```python
# Resize by ratio
width_ratio = 0.75
height_ratio = 0.75
image_1_resize_ratio = cv2.resize(image_1_rgb,(0,0),image_1_rgb,width_ratio,height_ratio)
plt.imshow(image_1_resize_ratio)
```

```python
flip_image_1 = cv2.flip(image_1_rgb,0)
plt.imshow(flip_image_1)
```
