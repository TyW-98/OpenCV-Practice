```python
# Read grayscale image in by adding 0 at the end

img2 = cv2.imread("DATA/rainbow.jpg",0)
plt.imshow(img2,cmap="gray")
```

```python
# If the value is greater than the threshold (127) it will be replace with 255 and if it below the threshold (127) it will be replace with 0. 

_, img_thresh = cv2.threshold(img2,127,255,cv2.THRESH_BINARY)
img_thresh
```

```python
plt.imshow(img_thresh,cmap="gray")
```

```python
# Value above threshold (127) it will be replaced with 0 and if below threshold (127) it will be replaced with value 255.

_, img_thresh = cv2.threshold(img2,127,255,cv2.THRESH_BINARY_INV)
img_thresh
plt.imshow(img_thresh,cmap="gray")
```

```python
# If value greater than threshold (127) it will replace it with 255. If it is below the threshold it will keep the original value.

_, img_thresh = cv2.threshold(img2,127,255,cv2.THRESH_TRUNC)
img_thresh
plt.imshow(img_thresh,cmap="gray")
```

```python
img1 = cv2.imread("DATA/crossword.jpg",0)
plt.imshow(img1,cmap = "gray")
```

```python
def enlarge_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = "gray")
```

```python
enlarge_pic(img1)
```

```python
_ , cw_thres = cv2.threshold(img1,188,255,cv2.THRESH_BINARY)
enlarge_pic(cw_thres)
```

```python
cw_thres_2 = cv2.adaptiveThreshold(img1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
enlarge_pic(cw_thres_2)
```

```python
blended_image = cv2.addWeighted(cw_thres_2,0.35,cw_thres,0.8,0.7)
enlarge_pic(blended_image)
```
